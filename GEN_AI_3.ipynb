```python
from langchain_community.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import os

# ✅ Set OpenRouter API key directly
os.environ["OPENAI_API_KEY"] = "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"  # 🔥 Replace with your OpenRouter API key

# ✅ Initialize the chatbot with Mixtral model
chat = ChatOpenAI(
    temperature=0.3,                                # Low temperature for accuracy
    model_name="mistralai/mixtral-8x7b",             # 🔥 Use Mixtral model
    openai_api_base="https://openrouter.ai/api/v1"
)

# ✅ Improved prompt template
template = """
You are a precise and professional translator.
Translate the given text from {source_language} to {target_language}.
Provide only the translated text, without repeating the input.

### Input:
{input_text}

### Output:
"""

prompt = PromptTemplate(
    input_variables=["source_language", "target_language", "input_text"],
    template=template
)

chain = LLMChain(llm=chat, prompt=prompt)

# ✅ Translation loop
print("🌍 Welcome to the Language Translator! Type 'exit' to quit.")
while True:
    source_language = input("Enter source language (e.g., English): ")
    if source_language.lower() == "exit":
        print("👋 Goodbye!")
        break

    target_language = input("Enter target language (e.g., Tamil): ")
    text = input(f"Enter the text to translate from {source_language} to {target_language}: ")

    response = chain.run({
        "source_language": source_language,
        "target_language": target_language,
        "input_text": text
    })

    print("\n🗣️ **Translated Text:**\n")
    print(response)
```

```python
from langchain.chat_models import ChatOpenAI
import os

# Set OpenRouter API Key
os.environ["OPENROUTER_API_KEY"] = "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"  # Replace with your OpenRouter API key

# OpenRouter Base URL
openrouter_base_url = "https://openrouter.ai/api/v1"

# Initialize OpenRouter LLM Model
chat_model = ChatOpenAI(
    model_name="google/gemini-pro",  # Use OpenRouter's GPT-4 Turbo or another model
    temperature=0.3,
    openai_api_key=os.getenv("OPENROUTER_API_KEY"),
    openai_api_base=openrouter_base_url
)

def translate_text(text, target_language):
    """Function to translate text into a target language."""
    prompt = f"""
    Translate the following text into {target_language}:

    Text: {text}

    Translation:
    """
    response = chat_model.predict(prompt)
    return response

# Example usage
text_to_translate = "come here Jaiprakash!!!"
target_language = "tamil"  # Change to any language you want

translation = translate_text(text_to_translate, target_language)
print(f"Translated Text ({target_language}):\n", translation)
```

```python
# Install needed packages (uncomment if not already installed)
# !pip install diffusers transformers torch moviepy

from diffusers import StableDiffusionPipeline
import torch
from moviepy.editor import ImageSequenceClip
import os

# Step 1: Load Stable Diffusion (CPU version)
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float32  # float32 for CPU
).to("cpu")

# Step 2: Generate images from prompt
prompt = "A futuristic city skyline at sunset, ultra realistic"
output_dir = "video_frames"
os.makedirs(output_dir, exist_ok=True)

num_frames = 3  # reduce for faster testing

print("Generating frames...")
for i in range(num_frames):
    image = pipe(prompt).images[0]
    image_path = os.path.join(output_dir, f"frame_{i:03d}.png")
    image.save(image_path)
    print(f"Saved frame {i + 1}/{num_frames}")

# Step 3: Combine images into a video
print("Creating video...")
clip = ImageSequenceClip(output_dir, fps=4)  # You can increase FPS
clip.write_videofile("generated_video.mp4", codec="libx264")
print("Video saved as generated_video.mp4")
```
