# **PDF Query Answering Chatbot**

```python
import PyPDF2
from transformers import pipeline

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
        return text

# Function to initialize the question-answering model
def get_answer(question, context):
    # Load a pre-trained QA model and tokenizer from Hugging Face
    qa_pipeline = pipeline("question-answering")

    # Provide context (the extracted PDF text) and the user's question
    result = qa_pipeline(question=question, context=context)
    return result['answer']

def chatbot(pdf_path):
    # Extract text from the provided PDF
    context = extract_text_from_pdf(pdf_path)

    print("Hello! I'm your PDF query-answering chatbot. Ask me anything related to the document!")
    print("Type 'exit' to quit.")

    while True:
        query = input("\nWhat is your question? ")
        if query.lower() == 'exit':
            print("Goodbye!")
            break
        else:
            answer = get_answer(query, context)
            print("\nAnswer:", answer)

# Example Usage
# Provide the path to your PDF document here
pdf_path =r"/content/python.pdf"
 # Change this to your PDF file path
chatbot(pdf_path)
```
